```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
#  Excel 
df = pd.read_excel('E2_8.xlsx', sheet_name='Sheet1')
#  5 
print(df.head())
#  'year' 
df = df.set_index('year')

# 
plt.plot(df)
plt.title('Time Series Plot')
plt.show()

# 
rolling_mean = df.rolling(window=12).mean()
rolling_std = df.rolling(window=12).std()

# 
plt.plot(df, label='Original')
plt.plot(rolling_mean, color='red', label='Rolling Mean')
plt.plot(rolling_std, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation')
plt.show()

#  Dickey-Fuller 
result = adfuller(df['wl'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

#  Ljung-Box 
lag_num = 12  # 
result = acorr_ljungbox(df['wl'], lags=lag_num)
print('Ljung-Box Test:')
for i in range(lag_num):
    print('Lag %d: %f' % (i+1, result[1][i]))
```

```python
# 
diff_df = df.diff().dropna()

# 
plt.plot(diff_df)
plt.title('Time Series Plot (1st difference)')
plt.show()

# 
rolling_mean = diff_df.rolling(window=12).mean()
rolling_std = diff_df.rolling(window=12).std()

# 
plt.plot(diff_df, label='Original')
plt.plot(rolling_mean, color='red', label='Rolling Mean')
plt.plot(rolling_std, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation (1st difference)')
plt.show()

#  Dickey-Fuller 
result = adfuller(diff_df['wl'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Read in the data
df = pd.read_excel('E2_8.xlsx', sheet_name='Sheet1')

# Set the year column as the index
df = df.set_index('year')

# Plot the time series
plt.plot(df)
plt.title('Water Level Time Series')
plt.xlabel('Year')
plt.ylabel('Water Level')
plt.show()

# Perform the Dickey-Fuller test to check for stationarity
result = adfuller(df['wl'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
mean = df['wl'].mean()
std = df['wl'].std()

print("Mean: ", mean)
print("Standard Deviation: ", std)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 
plot_acf(df['wl'], lags=30)
plot_pacf(df['wl'], lags=30)

plt.show()
```
